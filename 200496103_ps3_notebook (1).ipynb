 Step 4. Year 2061 is seemingly improper. Convert every year which are < 70 to 19XX instead of 20XX

Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days

```python
def stat(x):
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
newData.apply(stat)
```

### Question 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!


```python
df8 = pd.read_csv('actor_kill_counts.csv').sort_values(by="Count", ascending=True)
actor_names = df8['Actor'].values
kill_counts = df8['Count'].values

df8
```

```python
# plot lines
fig, ax = plt.subplots(figsize=(22,10))
bars = ax.barh( actor_names,kill_counts )
i=0
for bar in bars:
    width = bar.get_width() 
    label_y_pos = bar.get_y() + bar.get_height() / 2
    ax.text(width, label_y_pos, s=f'{actor_names[i]}', va='center')
    i=i+1

plt.legend()
plt.xlabel("Kill Counts")
plt.ylabel('Actor')
plt.show()
```
